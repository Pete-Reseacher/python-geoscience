BUILD A SUBSURFACE RESERVOIR QUALITY TRACK

```python
por_cutoff = 0.20

plt.figure()

plt.plot(df1["CPHI"], df1["DEPTH"])
plt.axvline(por_cutoff, linestyle="--", linewidth = 1.5, color = "red")

plt.fill_betweenx(df1["DEPTH"], por_cutoff,
                  df1["CPHI"], where=(df1["CPHI"] >= por_cutoff), color="yellow")
plt.gca().invert_yaxis()

plt.title("Reservoir Quality Track")
plt.xlabel("Porosity v/v")
plt.ylabel("Depth (ft)")
plt.show
```

Teacher note: "Fixes out of class"
Someone asked why there were gaps inbetween the colour-fillin areas above the porosity cutoff point.
Reasons include:
* NaN values
* Strings

We can thereby fix as thus:

```python
#First chech for NaN values
df1[["CPHI", "DEPTH"]].info()
df1[["CPHI", "DEPTH"]].isna().sum()
```

```python
#Next we can decide to convert to numpy arrays as taught in last week class
#reason because fillbetween() works well with numpy arrays

DEPTH = df1["DEPTH"].values.astype(float)
CPHI  = df1["CPHI"].values.astype(float)

```

```python
#then replot our data

plt.figure(figsize=(6,10))

plt.plot(CPHI, DEPTH)
plt.axvline(por_cutoff, linestyle="--", linewidth=1.5, color="red")

plt.fill_betweenx(DEPTH, por_cutoff, CPHI,
    where=(CPHI >= por_cutoff), color="yellow", alpha=0.5,
    interpolate=True   #Now when using arrays as we converted in the previous cell, it is important to interpolate
                  #this is because there may be some values that do not have a clear transition profile from low to high
                  #interpolate is a data error handling method that fills that gap in the porosity values for good filling.
)

plt.gca().invert_yaxis()

plt.title("Reservoir Quality Track")
plt.xlabel("Porosity v/v")
plt.ylabel("Depth (ft)")
plt.show()
```
