### SPECIAL EXERCISE

#### Groundwater Safety Classification

<p>You are analyzing groundwater samples for nitrate concentration (mg/L).</p>

<p>Environmental guidelines classify groundwater safety into three categories:</p>

* "Safe"

* "Caution"

* "Unsafe"

<p>Write a function called classify_groundwater() that takes two parameters:</p>

* nitrate (a number, in mg/L)

* agricultural_zone (a boolean: True or False)

<p>Classification Rules

If nitrate concentration is 40 mg/L or less, return "Safe"

If nitrate concentration is between 41 and 50 mg/L (inclusive), return "Caution"

If nitrate concentration is greater than 50 mg/L, return "Unsafe"

Special Condition (Important):
If the sample is from an agricultural zone (agricultural_zone == True), the acceptable nitrate limits are 5 mg/L higher in all categories.</p>

Your Task

Write a function that returns the correct groundwater safety classification based on the rules above.

```python
def Classify_groundwater(nitrate, agricultural_zone):
  threshold = 5 if agricultural_zone else 0
  if nitrate <= 40 + threshold:
    return "Safe"
  elif 41 + threshold <= nitrate <= 50 + threshold:
      return "Caution"
  else:
    return "unsafe"


# Example usage
print(Classify_groundwater(42, False))  #Caution
print(Classify_groundwater(45, True)) #Safe
```
